```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python

df = pd.read_csv('/content/hiring.csv')
df
```

```python
from word2number import w2n
```

```python
w2n.word_to_num(df['experience'][5])
```

```python
# prompt: choose the df experice column and print all the value and use for loop and print each value

for i, exp in enumerate(df['experience']):
  if not pd.isna(exp):
    df['experience'][i] = w2n.word_to_num(df['experience'][i])
```

```python
df
```

```python
df['experience'] = df.experience.fillna(0)
```

```python
df
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median())
```

```python
df
```

```python
#modell
model = linear_model.LinearRegression()
model.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
model.predict([[2,9,6]])
```

```python
plt.scatter(df['experience'],df['salary($)'])
```
