```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/carprices.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df['Car Model'].unique()
```

```python
# prompt: get dummies for car model in int

dummies = pd.get_dummies(df['Car Model'], dtype=int)
dummies.head()
```

```python
mergedf = pd.concat([df, dummies], axis='columns')
mergedf.head()
```

```python
X = mergedf.drop(['Car Model', 'Sell Price($)'], axis='columns')
X.head()
```

```python
Y = mergedf['Sell Price($)']
Y.head()
```

```python
#model
model = LinearRegression()
model.fit(X, Y)
```

```python
model.score(X,Y)
```

```python
model.predict([[69000,	6,	0,	1,	0]])
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
carle = le.fit_transform(df['Car Model'])
carle
```

```python
df1 = df
df1['Car Model'] = carle
df1

dfX = df1[['Car Model','Mileage','Age(yrs)']].values
dfX
```

```python
testX = df1.drop(['Sell Price($)'], axis='columns').values
testX
dfY = df1['Sell Price($)']
dfY
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('Car Model', OneHotEncoder(), [0])], remainder='passthrough')
ct = ct.fit_transform(dfX)
ct
```

```python
df1['Mileage'].unique()
```

```python
#model
nm = LinearRegression()
nm.fit(ct, dfY)
```

```python
nm.predict([[0,0,1,45000,4]])
```

```python
nm.predict([[0,1,0,86000,7]])
```

```python
nm.score(ct,dfY)
```
