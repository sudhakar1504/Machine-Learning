```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#load dataset
df = pd.read_csv('/content/canada_per_capita_income.csv')
df.head()
```

```python
#scatter plot
plt.scatter(df['year'],df['per capita income (US$)'],marker='+',color='red')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.title('Canada Per Capita Income')
plt.show()
```

```python
#model
model =linear_model.LinearRegression()
model.fit(df[['year']],df['per capita income (US$)'])
```

```python
# prompt: coefficient fornula or linear regression

# linear regression coefficient formula (m = covariance(x,y) / variance(x))
# In terms of the data:
# m = sum((xi - mean(x)) * (yi - mean(y))) / sum((xi - mean(x))^2)
# or a simplified version:
# m = (n * sum(xi*yi) - sum(xi)*sum(yi)) / (n*sum(xi^2) - (sum(xi))^2)
# and the intercept (b = mean(y) - m * mean(x))

# Let's calculate these using the pandas DataFrame

# Calculate the mean of year and per capita income
mean_year = df['year'].mean()
mean_income = df['per capita income (US$)'].mean()

# Calculate the numerator and denominator for the coefficient (m)
numerator = sum((df['year'] - mean_year) * (df['per capita income (US$)'] - mean_income))
denominator = sum((df['year'] - mean_year)**2)

# Calculate the coefficient (m)
calculated_coef = numerator / denominator

# Calculate the intercept (b)
calculated_intercept = mean_income - calculated_coef * mean_year

print(f"Calculated Coefficient (m): {calculated_coef}")
print(f"Calculated Intercept (b): {calculated_intercept}")

# Compare with the model's values
print(f"Model's Coefficient (m): {model.coef_[0]}")
print(f"Model's Intercept (b): {model.intercept_}")

# Note that the formulas are equivalent and should produce the same results as the scikit-learn model.
# The last line in the original code `5576.514583- 4804.463248/1974 -1973` seems incomplete or incorrect
# and is not a standard linear regression formula.
```
