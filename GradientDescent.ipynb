```python
# Gradient Descent
def gradientDescent(x,y):
  m=b=0
  LR = 0.0002
  iteration = 100000
  n = len(x)
  print(n)
  for i in range(iteration):
    y_predictt =m*x+b   # mx+b
    mse = (1/n) * sum([value**2 for value in (y-y_predictt)])
    dm = -(2/n) * sum(x*(y-y_predictt))     #-(2/n)*sum(x*(y-y_predictt))
    db = -(2/n) * sum((y-y_predictt)) #-(2/n) * sum((y-Y-predictt))
    m = m-LR*dm
    b = b-LR*db
    print(m,b,mse)

x = np.array(df.math)
y = np.array(df.cs)
print(y)
gradientDescent(df['math'],df['cs'])
```
